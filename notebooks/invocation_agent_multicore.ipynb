```python
import boto3
import json
import uuid

# Create AgentCore client
client = boto3.client(
    "bedrock-agentcore",
    region_name="us-east-1"
)

# Payload sent to your hosted agent
payload = {
    "prompt": "Explain machine learning in simple terms"
}

# Session IDs must be >= 33 chars
session_id = f"{uuid.uuid4()}-session-runtime"

# Invoke runtime
response = client.invoke_agent_runtime(
    agentRuntimeArn=(
        "arn:aws:bedrock-agentcore:us-east-1:"
        "014498646416:runtime/"
        "tom_hackathon-ChO02O61W2"
    ),
    runtimeSessionId=session_id,
    payload=json.dumps(payload),

    # Optional:
    # omit entirely if using default endpoint
    qualifier="DEFAULT"
)

# Read streaming body
raw_response = response["response"].read()

# Decode bytes
decoded_response = raw_response.decode("utf-8")

print("Raw Response:")
print(decoded_response)

# Try parsing JSON if agent returns JSON
try:
    parsed = json.loads(decoded_response)

    print("\nParsed Response:")
    print(json.dumps(parsed, indent=2))

except json.JSONDecodeError:
    print("\nResponse was not JSON.")
```
